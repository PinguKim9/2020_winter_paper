```python
import os, glob
import os.path
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import zipfile
```

```python
direc = r"all korea"
```

```python
files = os.listdir(direc)
```

```python
def get_stock(file_name) :
    df = pd.read_csv(str('all korea\\' + file_name))
    
    df['date'] = pd.to_datetime(df['날짜'].astype('str'), format='%Y/%m/%d')
    df['year'] = pd.DatetimeIndex(df['date']).year
    df['month'] = pd.DatetimeIndex(df['date']).month
    df['증감률'] = (df['시가'] - df['종가'])/df['종가']
    df['증감'] = [1 if x>0 else 0 if x==0 else -1 for x in df['증감률']]
    
    df = df.iloc[:, 2:]
    return df
```

```python
def stock_ud(file_name1, year, *arg, **key) :
    df = get_stock(file_name1)
    df_year = df[df['year']==year]
    
    time = df_year.loc[:, 'date']
    price = df_year.loc[:, '증감률']
    
    plt.plot(time, price, *arg, **key)
    plt.grid()
    
    return df_year
```

```python
def sim(file_name1, file_name2, year, day) :
    df1 = stock_ud(file_name1, year, "r-", label=str(file_name1))
    df2 = stock_ud(file_name2, year, "b-", label=str(file_name2))
    
    num = df1.shape[0]
    df_sim = pd.DataFrame(index=range(num), columns=['date', 'similarity'])
    
    for i in range(num) :
        df_sim['date'][num-1 - i] = df1['date'].iloc[i]
        
    try : 
        for i in range(num-1) :
            if df1['증감'].iloc[num-1-i -day] == df2['증감'].iloc[num-1-i] :
                df_sim['similarity'][i] = 1
            else : df_sim['similarity'][i] = 0
    except IndexError :
        print(file_name2)
        
    print('유사도 : ', df_sim['similarity'].mean() *100)        
    return df_sim
```

```python
for j in range(len(files)) :
        sim('삼성전자.csv', files[j], 2020, 2)
```
