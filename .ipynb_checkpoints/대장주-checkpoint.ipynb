```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import zipfile
```

```python
# 폰트 설정
plt.rc('font', family='NanumGothicCoding')
plt.rc('axes', unicode_minus=False) 
plt.rcParams["figure.figsize"] = (18,4)
%config InlineBackend.figure_format = 'retina'
```

```python
def get_stock(file_name) :
    z = zipfile.ZipFile('all korea.zip')
    file = str(file_name + '.csv')
    f = z.open(file)
    df = pd.read_csv(f)
    
    df['date'] = pd.to_datetime(df['날짜'].astype('str'), format='%Y/%m/%d')
    df['year'] = pd.DatetimeIndex(df['date']).year
    df['month'] = pd.DatetimeIndex(df['date']).month
    df['증감률'] = (df['시가'] - df['종가'])/df['종가']
    df['증감'] = [1 if x>0 else 0 if x==0 else -1 for x in df['증감률']]
    
    df = df.iloc[:, 2:]
    return df
```

```python
get_stock('삼성전자').head()
```

```python
def stock(file_name, *arg, **key) :
    df = get_stock(file_name)
    
    time = df.loc[:, 'date']
    price = df.loc[:, '종가']
    
    plt.plot(time, price, *arg, **key)
    plt.grid()
    
    return df
```

```python
stock('삼성전자').head()
```

```python
def stock2(file_name1, file_name2) :
    df1 = stock(file_name1, "r-", label=str(file_name1))
    df2 = stock(file_name2, "b-", label=str(file_name2))
    
    df1['name'] = file_name1
    df2['name'] = file_name2
    df3 = pd.concat([df1, df2])
    
    plt.legend()
    plt.grid()
    plt.show()
    
    return df3
```

```python
stock2("삼성전자", "SK하이닉스")
```

```python
def stock_year(file_name1, year, *arg, **key) :
    df = get_stock(file_name1)
    df_year = df[df['year']==year]
    
    time = df_year.loc[:, 'date']
    price = df_year.loc[:, '종가']
    
    plt.plot(time, price, *arg, **key)
    plt.grid()
    
    return df_year
```

```python
stock_year('삼성전자', 2020).head()
```

```python
def stock_year2(file_name1, file_name2, year) :
    df1 = stock_year(file_name1, year, "r-", label=str(file_name1))
    df2 = stock_year(file_name2, year, "b-", label=str(file_name2))
    
    df1['name'] = file_name1
    df2['name'] = file_name2
    df3 = pd.concat([df1, df2])
    
    plt.legend()
    plt.grid()
    plt.show()
    
    return df3
```

```python
stock_year2('삼성전자', 'SK하이닉스', 2020)
```

```python
def stock_ud(file_name1, year, *arg, **key) :
    df = get_stock(file_name1)
    df_year = df[df['year']==year]
    
    time = df_year.loc[:, 'date']
    price = df_year.loc[:, '증감률']
    
    plt.plot(time, price, *arg, **key)
    plt.grid()
    
    return df_year
```

```python
stock_ud('카카오', 2020).head()
```

```python
def stock_ud2(file_name1, file_name2, year) :
    df1 = stock_ud(file_name1, year, "r-", label=str(file_name1))
    df2 = stock_ud(file_name2, year, "b-", label=str(file_name2))
    
    df1['name'] = file_name1
    df2['name'] = file_name2
    df3 = pd.concat([df1, df2])
    
    plt.legend()
    plt.grid()
    plt.show()
    
    return df3
```

```python
stock_ud2('카카오', 'NAVER', 2020)
```

```python
def sim(file_name1, file_name2, year, day) :
    df1 = stock_ud(file_name1, year, "r-", label=str(file_name1))
    df2 = stock_ud(file_name2, year, "b-", label=str(file_name2))
    
    num = df1.shape[0]
    df_sim = pd.DataFrame(index=range(num), columns=['date', 'similarity'])
    
    for i in range(num) :
        df_sim['date'][num-1 - i] = df1['date'].iloc[i]
        
    for i in range(num-1) :
        if df1['증감'].iloc[num-1-i -day] == df2['증감'].iloc[num-1-i] :
            df_sim['similarity'][i] = 1
        else : df_sim['similarity'][i] = 0
        
    print('유사도 : ', df_sim['similarity'].mean() *100)        
    return df_sim
```

```python
sim('NAVER', '카카오', 2020, 0)
```

```python
for i in range(10) :
    sim('삼성전자', 'SK하이닉스', 2020, i)
```
