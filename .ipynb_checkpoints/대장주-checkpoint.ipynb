```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import zipfile
```

```python
def stock(file_name) :
    z = zipfile.ZipFile('all korea.zip')
    file = str(file_name + '.csv')
    f = z.open(file)
    df = pd.read_csv(f)
    print(df.head())
    
    df['date'] = pd.to_datetime(df['날짜'].astype('str'), format='%Y/%m/%d')
    time = df.loc[:, 'date']
    price = df.loc[:, '종가']
    plt.plot(time, price, '-')
    plt.grid()
```

```python
stock('삼성전자')
```

```python
stock('SK하이닉스')
```

```python
def stock_two(file_name1, file_name2) :
    z = zipfile.ZipFile('all korea.zip')
    
    file = str(file_name1 + '.csv')
    f = z.open(file)
    df = pd.read_csv(f)    
    df['date'] = pd.to_datetime(df['날짜'].astype('str'), format='%Y/%m/%d')
    
    time = df.loc[:, 'date']
    price = df.loc[:, '종가']
    plt.plot(time, price, 'r-')
    
    file = str(file_name2 + '.csv')
    f = z.open(file)
    df = pd.read_csv(f)    
    df['date'] = pd.to_datetime(df['날짜'].astype('str'), format='%Y/%m/%d')
    
    time = df.loc[:, 'date']
    price = df.loc[:, '종가']
    plt.plot(time, price, 'b-')
    
    plt.grid()
    plt.show()
```

```python
stock_two('삼성전자', 'SK하이닉스')
```

```python
def stock_year(file_name1, file_name2, year) :
    z = zipfile.ZipFile('all korea.zip')
    
    file = str(file_name1 + '.csv')
    f = z.open(file)
    df = pd.read_csv(f)    
    df['date'] = pd.to_datetime(df['날짜'].astype('str'), format='%Y/%m/%d')
    
    df['year'] = pd.DatetimeIndex(df['date']).year
    df_year = df[df['year']==year]
    
    time = df_year.loc[:, 'date']
    price = df_year.loc[:, '종가']
    plt.plot(time, price, 'r-')
    
    file = str(file_name2 + '.csv')
    f = z.open(file)
    df = pd.read_csv(f)    
    df['date'] = pd.to_datetime(df['날짜'].astype('str'), format='%Y/%m/%d')
    
    df['year'] = pd.DatetimeIndex(df['date']).year
    df_year = df[df['year']==year]    
    
    time = df_year.loc[:, 'date']
    price = df_year.loc[:, '종가']
    plt.plot(time, price, 'b-')
    
    plt.grid()
    plt.show()
```

```python
stock_year('삼성전자', 'SK하이닉스', 2020)
```

```python
def stock_month(file_name1, file_name2, year, month) :
    z = zipfile.ZipFile('all korea.zip')
    
    file = str(file_name1 + '.csv')
    f = z.open(file)
    df = pd.read_csv(f)    
    df['date'] = pd.to_datetime(df['날짜'].astype('str'), format='%Y/%m/%d')
    
    df['year'] = pd.DatetimeIndex(df['date']).year
    df_year = df[df['year']==year]
    
    df_year['month'] = pd.DatetimeIndex(df_year['date']).month
    df_month = df_year[df_year['month']==month]
    
    time = df_month.loc[:, 'date']
    price = df_month.loc[:, '종가']
    plt.plot(time, price, 'r-')
    
    file = str(file_name2 + '.csv')
    f = z.open(file)
    df = pd.read_csv(f)    
    df['date'] = pd.to_datetime(df['날짜'].astype('str'), format='%Y/%m/%d')
    
    df['year'] = pd.DatetimeIndex(df['date']).year
    df_year = df[df['year']==year]    
    
    df_year['month'] = pd.DatetimeIndex(df_year['date']).month
    df_month = df_year[df_year['month']==month]
    
    time = df_month.loc[:, 'date']
    price = df_month.loc[:, '종가']
    plt.plot(time, price, 'b-')
    
    plt.rcParams["figure.figsize"] = (20,4)
    plt.grid()
    plt.show()
```

```python
def stock2(file_name1, file_name2, year, month) :
    z = zipfile.ZipFile('all korea.zip')
    
    file = str(file_name1 + '.csv')
    f = z.open(file)
    df = pd.read_csv(f)    
    df['증감률'] = (df['시가'] - df['종가'])/df['종가']
    df['date'] = pd.to_datetime(df['날짜'].astype('str'), format='%Y/%m/%d')
    
    df['year'] = pd.DatetimeIndex(df['date']).year
    df_year = df[df['year']==year]
    
    df_year['month'] = pd.DatetimeIndex(df_year['date']).month
    df_month = df_year[df_year['month']==month]
    
    df_month['증감'] = df_month['증감률']
    
    for i in range(df_month.shape[0]) :
        if df_month['증감률'][i] > 0 :
            df_month['증감'][i] = 1
        else : df_month['증감'][i] = -1
    
    time = df_month.loc[:, 'date']
    price = df_month.loc[:, '증감률']
    plt.plot(time, price, 'r-')
    
    file = str(file_name2 + '.csv')
    f = z.open(file)
    df = pd.read_csv(f)    
    df['증감률'] = (df['시가'] - df['종가'])/df['종가']
    df['date'] = pd.to_datetime(df['날짜'].astype('str'), format='%Y/%m/%d')
    
    df['year'] = pd.DatetimeIndex(df['date']).year
    df_year = df[df['year']==year]    
    
    df_year['month'] = pd.DatetimeIndex(df_year['date']).month
    df_month = df_year[df_year['month']==month]
    
    df_month['증감'] = df_month['증감률']
    
    for i in range(df_month.shape[0]) :
        if df_month['증감률'][i] > 0 :
            df_month['증감'][i] = 1
        else : df_month['증감'][i] = -1

    
    time = df_month.loc[:, 'date']
    price = df_month.loc[:, '증감률']
    plt.plot(time, price, 'b-')
    
    plt.rcParams["figure.figsize"] = (20,4)
    plt.grid()
    plt.show()
```

```python
stock2('NAVER', '카카오', 2020, 10)
```
